```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
import requests

r = requests.get("https://kat.cr/hourlydump.txt.gz",stream=True)
with open("hourlydump.txt.gz","wb") as content:
    count=0
    
    for down_content in r.iter_content(1024):
        count+=1
        dotscount=count%5
        print("\r"+str(count)+" Kbs done "+"."*dotscount,end="")
        content.write(down_content)
```

```python
import gzip

file = gzip.open("hourlydump.txt.gz")
with open("hourlydump.txt","wb") as writefile:
    writefile.write(file.read())
```

```python
data = pd.read_table("hourlydump.txt",sep="|")
```

```python
data.columns = "torrent_info_hash|torrent_name|torrent_category|torrent_info_url|torrent_download_url|size|category_id|files_count|seeders|leechers|upload_date".split("|")
```

```python
all_hindi = data[data.torrent_name.str.contains("Hindi")]

DUBBED_MOVIES_CAT = [63,23,24]
all_hindi_movies = all_hindi[all_hindi.torrent_category == "Movies"]
all_hindi_movies[all_hindi_movies.category_id.isin(DUBBED_MOVIES_CAT)]
all_hindi_movies[["torrent_name","torrent_download_url"]]
# all_hindi_movies
```
